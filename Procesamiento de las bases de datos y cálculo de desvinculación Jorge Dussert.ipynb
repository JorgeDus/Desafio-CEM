### Número de estudiantes desvinculados del periodo 2022-2023. 

Revisemos cuales MRUNs estan en el dataframe de 2023 que no se encuentran en la matrícula 2023, para identificar estudiantes 
desvinculados. Con la función **set** tomaremos los registros únicos de 2022 y veremos cuales no están presentes en 2023.

```python
desvinculados = set(data_2022['MRUN']) - set(data_2023['MRUN'])
print(f'Existen {len(desvinculados)} estudiantes desvinculados nivel nacional.')
```

```python
# Check rápido para revisar si es que algun RUN de desvinculados está en el registro de 2023

data_2023[data_2023['MRUN']==desvinculados]
```

```python
# Creamos un dataframe con estudiantes desvinculados

df_desvinculados = data_2022[data_2022['MRUN'].isin(desvinculados)].reset_index(drop=True)
df_desvinculados.info()
```

Vemos que el dataframe contiene más registros que la cantidad de desvinculados que calculamos, esto es porque existen registros duplicados. Hay estudiantes que aparecen en más de un establecimiento. Para representar más fielmente los estudiantes desvinculados, vamos a eliminar duplicados, y quedarnos con el registro donde la columna ASISTENCIA tenga el mayor valor, y así eliminar aquellos registros repetidos donde la asistencia a un establecimiento es 0.

```python
# Vamos a revisar los registros duplicados, fijándonos en el MRUN, correspondiente a la máscara del RUN del alumno.

print(df_desvinculados.MRUN.duplicated())
print('-'*50)
print(f'Existen {df_desvinculados.MRUN.duplicated().sum()} registros duplicados')
```

```python
# Tomemos el último RUN del dataframe como ejemplo

ejemplorun = df_desvinculados['MRUN'].tail(1).values[0]
df_desvinculados[df_desvinculados['MRUN']==ejemplorun]
```

```python
# Agrupamos por MRUN y tomamos el máximo valor de ASISTENCIA para quedarnos solo con un registro por estudiante, y guardamos
# los indices

indices = df_desvinculados.groupby('MRUN')['ASISTENCIA'].idxmax().values

# Luego filtramos nuestro dataframe por los indices 

df_desvinculados_filtrado = df_desvinculados.iloc[indices].sort_values(by='RBD').reset_index(drop=True)
```

### Matrícula teórica y  tasa de desvinculación por región.

Para crear la siguiente tabla, en el dataframe de estudiantes desvinculados hacemos un **groupby** por región y contamos los registros, para tener los valores de desvinculados por región. 
Luego hacemos lo mismo pero con el dataframe de matrícula 2023 para tener los matriculados por región, y así calcular la matrícula teórica como aparece en el código. 
Con estos valores podemos calcular la tasa de deserción y crear nuestro dataframe con las columnas correspondientes, ordenadas en orden decreciente repecto a la tasas de desvinculación. 

```python
desvinculados_por_region = df_desvinculados_filtrado.groupby("NOM_REG_RBD_A").agg({'MRUN':'count'}).MRUN.values
matricula_por_region = data_2023.groupby('NOM_REG_RBD_A').agg({'MRUN':'count'}).MRUN.values
matricula_teorica_por_region = desvinculados_por_region + matricula_por_region
tasa_desvinculacion = np.round(desvinculados_por_region/matricula_teorica_por_region,3)
regiones = df_desvinculados.groupby("NOM_REG_RBD_A").agg({'MRUN':'count'}).MRUN.index

df_desvinculados_region = pd.DataFrame({'Region':regiones,
                                        'Desvinculados':desvinculados_por_region,
                                        'Matricula Teorica':matricula_teorica_por_region,
                                        'Tasa Desvinculacion':tasa_desvinculacion
                                       }).sort_values(by='Tasa Desvinculacion',ascending=False).reset_index(drop=True)
df_desvinculados_region
```

###  Matrícula teórica y tasa de desvinculación por tipo de jornada


Realizamos el mismo proceso que en el punto anterior, pero cambiando la variable a Tipo de Jornada, con el detalle de mappear los números de jornada con sus categorías correspondientes, para mejorar la legibilidad.

```python
desvinculados_por_jor = df_desvinculados_filtrado.groupby("COD_JOR").agg({'MRUN':'count'}).MRUN.values
matricula_por_jor = data_2023.groupby('COD_JOR').agg({'MRUN':'count'}).MRUN.values
matricula_teorica_por_jor = desvinculados_por_jor + matricula_por_jor
tasa_desvinculacion_jor = np.round(desvinculados_por_jor/matricula_teorica_por_jor,3)
jornadas = df_desvinculados_filtrado.groupby("COD_JOR").agg({'MRUN':'count'}).MRUN.index

df_desvinculados_jornada = pd.DataFrame({'Jornada':jornadas,
                                         'Desvinculados':desvinculados_por_jor,
                                         'Matricula Teorica':matricula_teorica_por_jor,
                                         'Tasa Desvinculacion':tasa_desvinculacion_jor
                                        }).sort_values(by='Tasa Desvinculacion',ascending=False).reset_index(drop=True)
df_desvinculados_jornada
```

```python
jornada_mapping = {1: 'Mañana', 2: 'Tarde', 3: 'Mañana y tarde',4: 'Vespertina / Nocturna'}
df_desvinculados_jornada['Jornada'] = df_desvinculados_jornada['Jornada'].map(jornada_mapping)

df_desvinculados_jornada
```
